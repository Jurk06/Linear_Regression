# Error in Linear Regression
*  Here we ll discuss with loss function 
*  Primary objective is to keep the loss function minimum.
*  Predicted value is closer to the actual values
*  Predicted values are denoted by y hat ().
*  Actual values are denoted by y.
*  Error = y - y hat  

# There are 4 type of error that we ll discuss here 
*  Mean absolute error (MAE)
*  Mean squared error (MSE)
*  Root mean square error (RMSE)
*  Root mean square log error (RMSLE)

# Mean Absolute error(MAE):-
*  Simple error 
*  MEA is the sum of absolute/possitive errors of all values
*  Find the difference between actual and predicted values
*  Take their possitive sum
*  Divide by the number of terms
*  get the mean values

# Mean Square Error (MSE)
* Also called MSD-Mean Square Deviation 
* Measures the average of squares of the error.
* Average squared difference between the estimated values and the actual value.
* It is always non-negative
* Closer to 1 is better 
* To minimize MSE, the model could be more accurate
* The model is closer to actual data
* One example of a linear regression using this method is the least squares method

# Root mean square error (RMSE):-
*  Also called Root Mean Squared Deviation
*  Square root of MSE yields root mean square error (RMSE).
*  Smaller is better
*  It is the standard deviation of error (residual error)
*   RMSE is sensitive to outliers
*  If there is a outlier it shows large values
