```python
X=train[['budget','popularity','runtime']]
y=train['revenue']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y, test_size=0.9)
```

```python
regg=LinearRegression()
```

```python
regg.fit(X_train, y_train)
```

```python
regg.coef_
```

```python
pred=regg.predict(X_test)
```

```python
plt.scatter(y_test, pred, color='red')
#plt.plot(X_test, pred, color='blue')
plt.xlabel("Revenue")
plt.ylabel("parametr")
plt.show()
```

```python
plt.plot(X_train,regg.predict(X_train))
```

```python
from sklearn import metrics
```

```python
metrics.mean_absolute_error(y_test, pred)
```

```python
metrics.mean_squared_error(y_test, pred)
```

```python
np.sqrt(metrics.mean_squared_error(y_test,pred))
```

```python
coeff=pd.DataFrame(regg.coef_, X.columns)
coeff.columns=["Coefficients"]
coeff
```
